```python
from source import sampling as spl
import sys
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', 500)
```

## Step 1 : Load the simulator

```python
model = spl.load_pgm()
nms = spl.get_var_nms()
```

## Step 2: Set the sample size

```python
sample_size = 100
df_sim = spl.random_sample(model, nms, sample_size)
df_sim.iloc[:10]
```

### Convert names to their IDs

```python
df_sim_num = spl.nam2num(df_sim)
df_sim_num.iloc[:10]
```

### Introduce missing values
- default missingness mechanism is Missing Cmpletely A Rndom (MCAR)
- mode: can be 'mcar', 'mar', or 'mnar'
- mcar_p: probability of having a missing value that is MCAR
- mar_p: probability of having a missing given different parent values. The first is the probability of having a missing value when its parent value is lower than threshold; The second is the probability of having a missing value when its parent value is higher than threshold.
- mnar_p: the same with mar_p


```python
df_sim_mcar = spl.add_missing_data(df_sim, mode='mcar', seed=10, mcar_p=0.1)
df_sim_mcar.iloc[:10]
```

### Introduce Selection Bias
- seed: used for randomly choosing selection variables
- prob: the probability of selection bias when the value selection variable is higher than the threshold


```python
df_sim_sb = spl.add_selection_bias(df_sim, prob=0.5, seed=10)
df_sim_sb.iloc[:10]
```

### Introduce unknown confounding

```python
df_sim_uc = spl.add_confounder(df_sim)
df_sim_uc.iloc[:10]
```
